```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/cars.csv')
df.head()
```

```python
car_features = df[['mpg','disp','hp']]
car_features.head()
```

```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wcss = []
for i in range(1,9):
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(car_features)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,9),wcss, marker='o')
plt.grid()
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit_predict(car_features)
clusters
```

```python
car_features['clusters']= clusters
```

```python
car_features
```

```python
kmeans = KMeans(n_clusters=5)
clusters = kmeans.fit_predict(car_features)
clusters
```

```python
car_features
```

```python
sns.scatterplot(data=car_features,x=car_features['mpg'],y=car_features['hp'],hue='clusters',palette='muted')
plt.show()
```
