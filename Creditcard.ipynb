```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/creditcard.csv.crdownload')
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df['Amount'].describe()
```

```python
non_fraud = len(df[df.Class == 0])
fraud = len(df[df.Class == 1])
fraud_percent = (fraud / (fraud + non_fraud)) * 100
print('No of Genuine transaction: ',non_fraud)
print('No of Fraud transaction: ',fraud)
print('percentage of fraud transaction: {:.4f}'.format(fraud_percent))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
labels = ['Genuine','Fraud']
count_classes = df.value_counts(df['Class'], sort=True)
count_classes.plot(kind= 'bar',rot=0)
plt.title('Visualization of Labels')
plt.ylabel('Count')
plt.xticks(range(2), labels)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['NormalizedAmount'] = scaler.fit_transform(df['Amount'].values.reshape(-1,1))
df.drop(['Amount','Time'],inplace=True,axis=1)
X = df.drop('Class',axis=1)
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print('Shape of X_train: ', X_train.shape)
print('Shape of y_test: ', y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
decision_tree_score = model.score(X_test,y_test) * 100
decision_tree_score
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
rf_pred = rf.predict(X_test)
rf_pred
```

```python
rf_score = rf.score(X_test,y_test) * 100
rf_score
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
```

```python
def metrics(actuals,predictions):
  print('Accuracy: {:.5f}'.format(accuracy_score(actuals,predictions)))
  print('Precision: {:.5f}'.format(precision_score(actuals,predictions)))
  print('Recall: {:.5f}'.format(recall_score(actuals,predictions)))
  print('F1-score: {:.5f}'.format(f1_score(actuals,predictions)))
```

```python
confusion_matrix(y_pred, y_test)
```
