```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/adult.csv')
df.head()
```

```python
df.isnull().sum().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['workclass'] = le.fit_transform(df['workclass'])
```

```python
df.head()
```

```python
df['education'] = le.fit_transform(df['education'])
```

```python
df.head()
```

```python
df['marital-status'] = le.fit_transform(df['marital-status'])
df['occupation'] = le.fit_transform(df['occupation'])
df['relationship'] = le.fit_transform(df['relationship'])
df['race'] = le.fit_transform(df['race'])
df['gender'] = le.fit_transform(df['gender'])
df['native-country'] = le.fit_transform(df['native-country'])
```

```python
df['income'] = le.fit_transform(df['income'])
```

```python
df.head()
```

```python
df.plot(kind='area')
plt.show()
```

```python
sns.barplot(data=df,x=df['age'],y=df['occupation'])
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(data=df,x=df['age'],y=df['workclass'])
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(data=df,x=df['age'],y=df['workclass'])
plt.show()
```

```python
sns.scatterplot(data=df,x=df['age'],y= df['fnlwgt'])
plt.show()
```

```python
corr = df.corr()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(corr, annot=True, fmt='0.1f',color='r')
plt.show()
```

```python
sns.scatterplot(data=df,x=df['gender'],y=df['income'])
plt.show()
```

```python
X = df.drop('income',axis=1)
```

```python
y = df['income']
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```
